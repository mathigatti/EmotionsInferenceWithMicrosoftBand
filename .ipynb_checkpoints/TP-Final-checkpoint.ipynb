```python
import csv
import os
labels = []
grs = []
temps = []
bpms = []
for file in os.listdir("Data"):
    if file == "eugeExcitacion1.txt" or file == "":
        with open("Data/" + file) as data:
            gr = []
            temp = []
            bpm = []
            spamreader = csv.reader(data, delimiter=',', quotechar='|')
            for row in spamreader:
                gr.append(float(row[0]))
                temp.append(float(row[1]))
                bpm.append(float(row[2]))
            grs.append(gr)
            temps.append(temp)
            bpms.append(bpm)
            labels.append(file)
```

```python
import pandas as pd

def toPlotableVector(signals):
    res = []
    for signal in signals:
        for i in range(len(signal)):
            if i > len(res) - 1:
                res.append([signal[i]])
            else:
                res[i].append(signal[i])
    return res

df1 = pd.DataFrame(toPlotableVector(grs))
df2 = pd.DataFrame(toPlotableVector(temps))
df3 = pd.DataFrame(toPlotableVector(bpms))
df1.columns = labels
df2.columns = labels
df3.columns = labels

%matplotlib inline
df1.plot()
df2.plot()
df3.plot()
```

```python
import pymc3 as pm, theano.tensor as tt

# setup model
model = pm.Model()
with model:

    p = pm.Uniform("p", 0, 1)
    activeness = pm.Bernoulli('activeness',p)

    mean_range = pm.Deterministic('mean_lower', pm.math.switch(activeness, [40,80], [60,160]))
    
    mean = pm.Uniform('mean', lower=mean_range[0], upper=mean_range[1])
    sd = pm.Uniform('sd', lower=0, upper=20)

    points1 = pm.Normal('obs1',
                       mu=mean,
                       sd=sd,
                       observed=bpms[1])
```

```python
# fit model
with model:
    step = pm.Metropolis()
    tr = pm.sample(10000, step=step)
```

```python
pm.plots.traceplot(tr, ['p','mean','sd']);
```
